```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy as np
```

```python
# x is Total rides for the city
# y is Average fare for city
# so each x,y pair is represents one city's data e.g. (50, 17)
x=[50,42,39,34,32,48,22,17,8,13]
y=[17,19,21,22,24,23,34,31,43,35]
```

```python
plt.scatter(x,y,c='green',marker='x',s=100)
plt.xlabel('Total number of Riders (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Average Fare vs Total Rides')
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis,c='red',label='Chicago')
plt.legend()
plt.ylabel('Total number of Riders (Per City)')
plt.xlabel('Average Fare ($)')
plt.title('Average Fare vs Total Rides')
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
y_axis_larger = []
#for data in y_axis:
#  y_axis_larger.append(data*3)
#plt.scatter(x_axis, y_axis, s=y_axis_larger)
#Use list comprehension
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Scatter plot using the object-oriented approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Bubble Chart using the object-oriented approach
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis
           , s=[i * 5 for i in y_axis]
           ,color='xkcd:sky blue'
           ,alpha=0.8
           ,edgecolor='black' 
           ,label='Boston'
           ,linewidth=2
           )
ax.legend()
plt.xlim(0, 50)
ax.set_title('Fares')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Month')
plt.gca().invert_yaxis()
```
